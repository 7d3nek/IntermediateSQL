```python
%%sql

WITH monthly_sales AS (
  SELECT
    TO_CHAR(orderdate, 'YYYY-MM') AS month,
    SUM(quantity * netprice * exchangerate) AS net_revenue
  FROM sales
  WHERE EXTRACT(YEAR FROM orderdate) = 2023
  GROUP BY month
  ORDER BY month
)
SELECT
  month,
  net_revenue,
  AVG(net_revenue) OVER (
    ORDER BY month
    ROWS BETWEEN CURRENT ROW AND CURRENT ROW
  ) AS net_revenue_current
FROM monthly_sales;
```

```python
%%sql
/*
Calculate a 7-day rolling revenue window to analyze short-term trends in daily sales.
This helps smooth out daily fluctuations and highlight patterns in weekly performance.

Use a CTE to calculate the total net revenue per day by grouping sales by orderdate.
In the main query, use window functions with a 7-day frame to calculate:
The rolling sum of revenue (3 days before through 3 days after)
The rolling average of revenue over the same window
This rolling view helps detect local spikes, dips, and trends in daily revenue.
*/
WITH daily_sales AS (
  SELECT
    orderdate,
    SUM(quantity * netprice * exchangerate) AS net_revenue
  FROM sales
  GROUP BY orderdate
  ORDER BY orderdate
)
SELECT
  orderdate,
  net_revenue,
  SUM(net_revenue) OVER(ORDER BY orderdate ROWS BETWEEN 3 PRECEDING AND 3 FOLLOWING),
  AVG(net_revenue) OVER(ORDER BY orderdate ROWS BETWEEN 3 PRECEDING AND 3 FOLLOWING)
FROM daily_sales
```

```python
%%sql
/*
Analyze short-term order revenue trends by calculating the total and average revenue of the next 10 orders following each order.
This helps identify periods of high or low revenue momentum based on future order values.

Use a CTE to calculate the net revenue per order by grouping on orderdate and orderkey.
In the main query, use window functions to calculate:
The rolling sum of revenue for the current and next 9 orders
The rolling average revenue over the same window
This gives a forward-looking view of how each order fits into the surrounding revenue stream.
*/

WITH daily_sales AS (
  SELECT
    orderdate,
    orderkey,
    SUM(quantity * netprice * exchangerate) AS net_revenue
  FROM sales
  GROUP BY orderdate, orderkey
  ORDER BY orderdate, orderkey
)
SELECT
  orderdate,
  orderkey,
  net_revenue,
  SUM(net_revenue) OVER(ORDER BY orderdate, orderkey ROWS BETWEEN CURRENT ROW AND 9 FOLLOWING),
  AVG(net_revenue) OVER(ORDER BY orderdate, orderkey ROWS BETWEEN CURRENT ROW AND 9 FOLLOWING
)
FROM daily_sales
```

```python
%%sql
/*
Analyze weekly revenue trends for the year 2022, and calculate cumulative, future, and specific week revenues to understand revenue performance across the entire year.

Use a CTE to compute total net revenue per week using TO_CHAR(orderdate, 'WW-YYYY').
In the main query, apply window functions to calculate:
The cumulative revenue up to the current week
The total future revenue from the current week forward
The revenue in the final week of the dataset
The revenue in the fifth week of the year
*/
WITH weekly_sales AS (
  SELECT
    TO_CHAR(orderdate, 'WW-YYYY') AS week,
    SUM(quantity * netprice * exchangerate) AS net_revenue
  FROM sales
  WHERE EXTRACT(YEAR FROM orderdate) = 2022
  GROUP BY week
  ORDER BY week
)
SELECT
  week,
  net_revenue,
  SUM(net_revenue) OVER(ORDER BY week ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS cumulative_revenue,
  SUM(net_revenue) OVER(ORDER BY week ROWS BETWEEN CURRENT ROW AND UNBOUNDED FOLLOWING) AS future_revenue,
  LAST_VALUE(net_revenue) OVER(ORDER BY week ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS last_value,
  NTH_VALUE(net_revenue, 5) OVER(ORDER BY week ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS fifth_week_revenue
FROM weekly_sales;
```
